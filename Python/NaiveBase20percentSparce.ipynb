```python
#Loading data set obama.csv
dataset = pd.read_csv("obama.csv", delimiter = ",")
```

```python
#View dataset
dataset
```

```python
# Separating Target and Independent variables
X = dataset.ix[:,1:32]
y = dataset.ix[:,32]
```

```python
# Splitting train and test data sets
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 21)
```

```python
#import and instantiate a mulit-nomial naive bayes model
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

```python
# train the model using X_train
nb.fit(X_train, y_train)
```

```python
# make class predictions for X_train
y_train_pred = nb.predict(X_train)
```

```python
# make class predictions for X_test
y_test_pred = nb.predict(X_test)
```

```python
# calculate Accuracy of class predictions
from sklearn import metrics
print(metrics.accuracy_score(y_train, y_train_pred))
print(metrics.accuracy_score(y_test, y_test_pred))
```
